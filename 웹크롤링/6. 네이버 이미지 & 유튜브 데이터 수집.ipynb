```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from tqdm.notebook import tqdm as tn
from bs4 import BeautifulSoup as bs
from urllib.request import urlretrieve

import pandas as pd
import time
```

## 네이버 이미지 수집

```python
url = 'https://search.naver.com/search.naver?where=image&sm=tab_jum&query=%EC%B9%B4%EC%B9%B4%EC%98%A4+%EB%9D%BC%EC%9D%B4%EC%96%B8'

driver = webdriver.Chrome()
driver.get(url)
```

```python
for i in range(15):
    driver.find_element_by_css_selector('body').send_keys(Keys.SPACE)
    time.sleep(1)
```

```python
html = bs(driver.page_source, 'html.parser')
driver.quit()
```

```python
# 이미지 요소들 수집
len(html.select('img._image._listImage'))
```

```python
img = html.select('img._image._listImage')
img
```

```python
src_list = []

for item in img:
    # 요소의 속성값 접근
    
    # src 속성이 없는 경우
    if item.get('data-lazy-src') != None:
        src_list.append(item.get('data-lazy-src'))
    else:
        src_list.append(item.get('src'))
        
src_list
```

```python
ext_list = []

for item in src_list:
    start = item.rfind('.')+1
    end = item.rfind('&')
    
    ext_list.append(item[start:end])
```

```python
for i in tn(range(len(ext_list))):
    urlretrieve(src_list[i], 'images/{}_{}.{}'.format('lion', i+1, ext_list[i]))
```

## 유튜브 데이터 수집
- 영상 제목, 영상 주소, 조회수, 섬네일 주소 수집
- 유튜브_영상정보.csv 파일로 저장

```python
url = 'https://www.youtube.com/results?search_query=%EA%B0%95%EC%95%84%EC%A7%80'

driver = webdriver.Chrome()
driver.get(url)
```

```python
for i in range(15):
    driver.find_element_by_css_selector('body').send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
```

```python
html = bs(driver.page_source, 'html.parser')
driver.quit()
```

```python
info = html.select('a#video-title')
thumbnail = html.select('#thumbnail #img')
```

```python
info
```

```python
title = []
link = []
view_cnt = []
thumb_src = []
thumb_ext = []

for i in range(len(info)):
    title.append(info[i].text.strip())
    
    link.append('http://www.youtube.com' + info[i].get('href'))
    
    start = info[i].get('aria-label').rfind('조회수') + 4
    view_cnt.append(info[i].get('aria-label')[start:-1].replace(',', ""))
    
title, link, view_cnt
```

```python
for i in range(len(thumbnail)):
        if thumbnail[i].get('src') != None:
            thumb_src.append(thumbnail[i].get('src'))
            start = thumbnail[i].get('src').rfind('.')+1
            end = thumbnail[i].get('src').find('?sqp')

            thumb_ext.append(thumbnail[i].get('src')[start:end])
```

```python
youtube_df = pd.DataFrame({"영상제목":title, "영상링크":link, "조회수":view_cnt, "썸네일 주소":thumb_src})
youtube_df
```

```python
youtube_df.to_csv('유튜브_영상정보.csv', encoding='utf-8-sig', index=None)
```
