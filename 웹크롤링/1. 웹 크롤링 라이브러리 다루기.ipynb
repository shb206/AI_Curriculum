# 웹 크롤링
- 웹 사이트의 정보(텍스트, 이미지, 동영상, ...)를 가져오고 저장하는 기술

## requests 라이브러리
- 접근할 웹 페이지를 요청/응답받기 위한 라이브러리

## BeautifulSoup
- 응답받은 데이터에서 원하는 데이터를 추출하기 위해 변환해주는 라이브러리

```python
import requests as req
from bs4 import BeautifulSoup as bs
```

```python
# 데이터를 요청하는 전송 방식(기본)
# get(url) : 해당 URL 주소를 요청/응답받는 함수
# 리턴타입 : Response 타입
res = req.get('http://www.google.co.kr')

# 요청 상태 정보 ex) 200 - 요청 성공, 400번대 - 요청을 잘못했을 때, 500번대 - 서버 문제
print(res.status_code)

# 응답받은 데이터 -> HTML 소스코드
print(res.text)
```

```python
url = 'https://www.google.co.kr/search?'

# params : 해당 url 주소로 요청을 보낼 때 값을 전달하는 속성(딕셔너리타입)
res = req.get(url, params={'q':'요가'})

print(res.text)
```

## HTML 소스코드(문자열) -> 파이썬 객체로 변환

```python
keyword = input('검색할 키워드 입력 : ')

url = 'https://search.naver.com/search.naver?'

# params : 해당 url 주소로 요청을 보낼 때 값을 전달하는 속성(딕셔너리타입)
res = req.get(url, params={'query':keyword})

# 파싱 : 웹 페이지 내 사용자가 원하는 데이터를 분석하여 변환 & 추출
# 파서(파싱 프로그램) : lxml, html.parser, html5lib
# bs(변환할 데이터, 파서)
html = bs(res.text, 'html.parser')

print(type(html))
```

```python
print(html)
```

```python
# 하나의 요소 접근
# 여러개가 있을 경우 가장 처음 것을 가져옴
# select_one(css 선택자)
print(html.select_one('.link_tit'))
```

```python
# 요소가 가진 내용 접근 : text 변수 사용
print((html.select_one('.link_tit').text.strip()))
```

```python
# 클래스가 여러개라면 각각의 클래스 앞에 .을 붙여줘야 함
# 여러 개의 요소에 접근
# select(css 선택자) : 리스트로 반환
# 리스트 타입의 경우 text 변수를 사용할 수 없음 -> 요소 하나하나에 접근해야 함
html.select('.api_txt_lines.total_tit')
```

```python
for view in html.select('.api_txt_lines.total_tit'):
    print(view.text)
```

## 네이버 뉴스 제목/내용 수집

- https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=421&aid=0005125859

```python
# 1. 해당 url 주소 요청/응답
url = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=023&aid=0003592008'

# 크롤링을 막아놓은 사이트도 존재
# 유저가 브라우저를 통해 접근하는거처럼 하는 방법
# 헤더값은 반드시 요청할 때 설정할 것
header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'
    # 그래도 안되면 + referer 속성 추가
}

res = req.get(url, headers=header)

print(res.status_code)
```

```python
# 2. bs 라이브러리로 응답받은 HTML 데이터 -> 파이썬 객체로 변환
html = bs(res.text, 'html.parser')
```

```python
# 웹 페이지의 모든 script 태그 접근
scripts = html.select('script')
# <script> 태그 제거
for script in scripts:
    script.extract()
    
# 이미지 설명 글 제거
emTags = html.select('em')
# <script> 태그 제거
for em in emTags:
    em.extract()

# 클래스로는 제대로 출력되지 않았음
# 웹페이지 동작과정에서 동적으로 생성되는 클래스는 select 메소드로는 확인이 불가능할 수 있음
title = html.select_one('#articleTitle').text
title
```

```python
# 3. select_one 함수를 이용하여 제목/내용을 가진 요소에 접근하고 text 요소에 접근
content = html.select_one('#articleBodyContents').text.strip()
content
```

```python
print('뉴스 제목 >> ', title)
print('뉴스 내용 >> ', content)
# 내용 수정
# print('뉴스 내용 >> ', content.replace('수정할 내용', ''))
```

## 멜론 일간 차트 1-100위 정보 수집
- https://www.melon.com/chart/day/index.htm
- 곡명, 가수, 앨범

```python
url = 'https://www.melon.com/chart/day/index.htm'

header = {
    'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.141 Safari/537.36'
}

res = req.get(url, headers=header)

html = bs(res.text, 'html.parser')
```

```python
song = html.select('div.ellipsis.rank01')
singer = html.select('div.ellipsis.rank02 > span')
album = html.select('div.ellipsis.rank03')
getImg = html.select('.image_typeAll > img')
```

### DataFrame 만드는 방법 - 1

```python
import pandas as pd
```

```python
rank_list = []
song_list = []
singer_list = []
album_list = []
img_list = []
i = 0
for i in range(len(singer)):
    rank_list.append(i+1)
    song_list.append(song[i].text.strip())
    singer_list.append(singer[i].text.strip())
    album_list.append(album[i].text.strip())
    img_list.append(getImg[i].get('src'))
    
len(song_list), len(singer_list), len(album_list), len(img_list)
```

```python
# for i in range(len(getImg)):
#     # 요소의 속성값 접근 : 오소.get(속성 이름) -- > 문자열로 반환
#     print('이미지주소 >> ', getImg[i].get('src')) 
```

```python
for i in range(len(song)):
    print("{}위 - {}, 가수 : {}, 앨범 : {}".format(i+1, song[i].text.strip(), singer[i].text.strip(), album[i].text.strip()))
    print("이미지 주소 >> ", getImg[i].get('src'))
```

```python
data = { "순위" : rank_list,
        "곡명" : song_list,
        "가수" : singer_list,
        "앨범" : album_list,
        "이미지주소" : img_list
       }

df = pd.DataFrame(data)
df.set_index('순위', inplace = True)
df
```

```python
df.to_csv('melon_music.csv', encoding='utf-8-sig')
```
