```python
import matplotlib.pyplot as plt
```

```python
y = [2, 4, 6]
x = [3, 6, 5]

# plot() : line chart를 그리는 메소드
# 매개변수가 하나면 y축을 기준으로 함, x축은 없다면 데이터의 갯수를 0부터
plt.plot(x, y)
plt.show()
```

```python
plt.plot(y, ls = "-.")
plt.show()
```

```python
plt.plot(y, marker = 'D')
plt.show()
```

```python
plt.plot(y, color='y')
plt.show
```

```python
plt.plot(y, color = '#00FF00')
plt.show()
```

```python
plt.plot(y)
plt.xlim(0, 10) # xlim(시작, 끝) : x축 범위
plt.ylim(2, 50) # ylim(시작, 끝) : y축 범위
plt.show()
```

```python
plt.plot(y)
plt.xticks([0, 3, 6]) # x 축에서 기준으로 보여줄 값
plt.yticks([1, 5, 9]) # y 축에서 기준으로 보여줄 값
plt.grid() # tick에 맞춰 배경에 선을 그어 줌
plt.show()
```

```python
x = [1, 2, 3, 4]
y = [3, 4, 5, 7]
z = [2, 4, 8, 16]
plt.plot(x)
plt.plot(y, ls = "--")
plt.plot(z, ls = "-.")
plt.show()
```

```python
plt.plot(x)
plt.plot(y)
plt.show()

plt.plot(z)
plt.show()
```

```python
plt.plot(y, label = "y") # 데이터에 라벨 추가
plt.legend() # 범례 출력
plt.show()
```
